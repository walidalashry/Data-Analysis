## Investigation Overview

> In this invistigation, I want to know what are the stations that does not have bikes at certain times of day. The main focus is the stations and the time. 

## Dataset Overview

>This data consists of 183412 bike trips in Feb-2019. it provides information about start and end time of each trip, duration, start and end station for each trip, start and end station id for each trip, start and end station latitude and longitude, bike_id, user's birthday and gender, customer type and bike share information. 

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('gobike.csv')
top_3_stations = pd.read_csv('top_3_stations.csv',index_col='start_time')
df.info()
```

## (Visualization 1)

> The distibution of trips in workdays are bimodal(at the going to and returning from work hours) while the distribution of trips in weekend are nearly normal.

```python

#convert date from string to datetime 
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

# Extract weekday and hour 
df['weekday'] = df.start_time.dt.weekday
df['hour'] = df.start_time.dt.hour
#divide df to workdays and weekend:
df_workday = df.query('weekday!=5 & weekday!=6')
df_weekend = df.query('weekday==5 | weekday==6')
#ploting 
base_color = sb.color_palette()[0]

plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
sb.countplot(df_workday.start_time.dt.hour, color=base_color)
plt.ylabel('number of trips')
plt.title('Number of trips during the hours of the day for workdays')
plt.xlabel('hour')

plt.subplot(1,2,2)
sb.countplot(df_weekend.start_time.dt.hour, color=base_color)
plt.ylabel('number of trips')
plt.title('Number of trips during the hours of the day for weekends');
plt.xlabel('hour');
```

## (Visualization 2)

> The number of trips in workdays are relatively close to each other ,then the number of trips decreased dramatically by around 50% at weekends.

```python

days = ['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su']
dayclasses = pd.api.types.CategoricalDtype(ordered=True, categories=days)
df.replace({'weekday': {0: 'Mo', 1: 'Tu',2: 'We',3: 'Th',4: 'Fr',5: 'Sa',6: 'Su'}}, inplace =True)
df.weekday = df.weekday.astype(dayclasses)

trips_count = df.weekday.value_counts().reindex(days)

plt.figure(figsize=(15,8))
sb.countplot(df.weekday, color=base_color)
plt.ylabel('number of Trips')
plt.title('Cumulative Number of trips at each weekday')

locs, labels = plt.xticks(rotation=15) 

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = trips_count[label.get_text()]
    pct_string = count

    # print the annotation just below the top of the bar
    plt.text(loc, count+200, pct_string, ha = 'center', color = 'black')
```

## (Visualization 3)

> Most of our trips are 30 minutes or less with average of 12 minutes. There are some trips with very high duration time. 


```python

log_binsize = 0.025
bins = 10 ** np.arange(1.78, np.log10(df.duration_sec.max())+log_binsize, log_binsize)

plt.figure(figsize=[15, 8])
plt.hist(data = df, x = 'duration_sec', bins = bins)
plt.xscale('log')
plt.xticks([50,100,200,500, 1e3, 2e3, 5e3, 1e4, 2e4,1e5], 
           [50,100,200,500, '1000', '2000', '5000', '10000', '20000','100000'])
plt.xlabel('duration(sec)')
plt.ylabel('frequancy');
plt.title('duration distribution at log scale');
```

## (Visualization 4)

> There are 15097 trips that starts from station that is not the same station that the same bike ended at which means the company tranfer them or people using bikes without starting a trip.

```python


df_ordered = pd.read_csv('gobike_ordered.csv')
df_ordered.drop(columns=['Unnamed: 0','index'], inplace=True)
#some type convertions
df_ordered['start_time'] = pd.to_datetime(df_ordered['start_time'])
df_ordered['end_time'] = pd.to_datetime(df_ordered['end_time'])
df_ordered['start_station_id'] = df_ordered['start_station_id'].astype(int).astype(str)
df_ordered['end_station_id'] = df_ordered['end_station_id'].astype(int).astype(str)
df_ordered['bike_id'] = df_ordered['bike_id'].astype(int).astype(str)

#modify the first bike id as 'new':
df_ordered.loc[0,'transfered'] = 'new'

df_ordered.transfered.replace({'new':'Not Transferred', 'False':'Not Transferred', 'True':'Transferred'}, inplace=True)

trans_count = df_ordered.transfered.value_counts()

plt.figure(figsize=(6,10))
sb.countplot(df_ordered['transfered'])
plt.title('% transferred trips')
plt.ylabel('Number of trips')
locs, labels = plt.xticks() 

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = trans_count[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/df_ordered.shape[0])

    # print the annotation just below the top of the bar
    plt.text(loc, count+2500, pct_string, ha = 'center', color = 'black')  
plt.legend(df_ordered['transfered'], loc=1)  
plt.show()
```
