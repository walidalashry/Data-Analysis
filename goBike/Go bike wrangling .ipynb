```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df_ordered = pd.read_csv('gobike_ordered.csv')
df_ordered.drop(columns=['Unnamed: 0','index'], inplace=True)
#some type convertions
df_ordered['start_time'] = pd.to_datetime(df_ordered['start_time'])
df_ordered['end_time'] = pd.to_datetime(df_ordered['end_time'])
df_ordered['start_station_id'] = df_ordered['start_station_id'].astype(int).astype(str)
df_ordered['end_station_id'] = df_ordered['end_station_id'].astype(int).astype(str)
df_ordered['bike_id'] = df_ordered['bike_id'].astype(int).astype(str)
```

```python
df_counts = df_ordered.copy()
df_counts['start_station_id'] = 's-'+ df_counts['start_station_id'].astype(str)
df_counts['end_station_id'] = 'f-'+ df_counts['end_station_id'].astype(str)
```
